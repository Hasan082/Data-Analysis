# Data Cleaning and Data Wrangling

### Finding Duplicates

```python
import pandas as pd
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"
df = pd.read_csv(path)
df.head()
```

### Task 1: Identify Duplicate Rows

```python
duplicated = df[df.duplicated()]
print(len(duplicated))
```

### Task 2: Analyze Characteristics of Duplicates

```python
duplicate_count = df.groupby(['MainBranch', 'Employment', 'RemoteWork']).size().reset_index(name='count')
duplicate_count
```

```python
frequency = duplicate_count[duplicate_count['count']> 1]
print(len(frequency))
```

#### Drop Duplicates

```python
df = df.drop_duplicates()
df.head()
```

```python
duplicated1 = df[df.duplicated()]
print(len(duplicated1))
```
