```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
```

```python
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can
```

```python
df_can.index.name = None
years = list(map(str, range(1980, 2014)))
years
```

```python
%matplotlib inline 
import matplotlib as mpl
mpl.style.use(['ggplot'])
```

# Line Pots

```python
# Extract data for Haiti
years = list(map(str, range(1980, 2014)))
haiti = df_can.loc['Haiti', years]
haiti
```

```python
haiti.plot(kind='line')
plt.title("Immigration from Haiti")
plt.xlabel("Year")
plt.ylabel("Immigration")
plt.show()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')
plt.title("Immigration from Haiti")
plt.xlabel("Year")
plt.ylabel("Immigration")
plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

# Question: Let's compare the number of immigrants from India and China from 1980 to 2013.

```python
df_CI = df_can.loc[['China', 'India'], years]
df_CI.head()
```

```python
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.plot(kind='line')
plt.title("Immigration from China and India")
plt.xlabel("Year")
plt.ylabel("Immigration")
plt.show()
```

# Question: Compare the trend of top 5 countries that contributed the most to immigration to Canada.

```python
inplace = True
df_can.sort_values(by="Total", ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)
df_top5
```

```python
df_top5 = df_can.sort_values(by="Total", ascending=False).head(5)  # Sort and get top 5
df_top5 = df_top5[years].transpose()
df_top5.plot(kind='bar', figsize=(14, 8))
plt.title("Immigration of Top 5 Countries")
plt.xlabel("Year")
plt.ylabel("Immigration")
plt.show()
```

```python
df_top5 = df_can.sort_values(by="Total", ascending=False).head(5)  # Sort and get top 5
df_top5 = df_top5[years].transpose()
df_top5.plot(kind='line', figsize=(14, 8))
plt.title("Immigration of Top 5 Countries")
plt.xlabel("Year")
plt.ylabel("Immigration")
plt.show()
```

```python
df_top5 = df_can.sort_values(by="Total", ascending=False).head(5)  # Sort and get top 5
df_top5 = df_top5[years].transpose()
df_top5.plot(kind='barh', figsize=(10, 10))
plt.title("Immigration of Top 5 Countries")
plt.xlabel("Year")
plt.ylabel("Immigration")
plt.show()
```

```python
df_top5 = df_can.sort_values(by="Total", ascending=False).head(5)  # Sort and get top 5
df_top5 = df_top5[years].transpose()
df_top5.plot(kind='area', figsize=(14, 8))
plt.title("Immigration of Top 5 Countries")
plt.xlabel("Year")
plt.ylabel("Immigration")
plt.show()
```

```python
df_top5 = df_can.sort_values(by="Total", ascending=False).head(5)
df_top5 = df_top5['Total']
df_top5.plot(kind='pie', figsize=(10, 10), autopct='%1.1f%%')
plt.title("Immigration of Top 5 Countries")
plt.show()
```
